```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("social_media_mental_health.csv")

df.head()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace("-", "_")
    .str.replace(" ", "_")
)
df.head()
```

```python
facebook_count = df[df['social_media_platform'] == 'Facebook']

media_tempo_gasto_facebook = facebook_count['daily_screen_time(hrs)'].mean()

print(f"A média de tempo gasto no Facebook é de {media_tempo_gasto_facebook:.1f} horas.")
```

```python
tiktok_count = df[df['social_media_platform'] == 'TikTok']

media_tempo_gasto_tiktok = tiktok_count['daily_screen_time(hrs)'].mean()

print(f"A média de tempo gasto no TikTok é de {media_tempo_gasto_tiktok:.1f} horas.")
```

```python
sns.lmplot(data=df, x="daily_screen_time(hrs)", y="stress_level(1_10)", scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Tempo de Tela x Estresse")
plt.xlabel('Tempo de Tela(hrs)', fontsize=12)
plt.ylabel('Nível de Estresse', fontsize=12)

plt.show()
```

```python
correlation = df["daily_screen_time(hrs)"].corr(df["stress_level(1_10)"])
print("Correlação entre tempo de tela e estresse:", round(correlation, 2))
```
